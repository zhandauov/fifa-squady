```python
import pandas as pd
#import numpy as np
players = pd.read_csv('../dataset light memory/players_fifa22.csv')
formation = ['GK', 
         'LB', 'CB', 'CB', 'RB', 
         'LM', 'CAM', 'RM',
         'LW', 'ST', 'RW']
money = 100000000000
one_player_balance = money / 11
```

```python
players.columns
```

```python
players
```

```python
squad = []
ids = []
ilc = 0
used_flag = False
for pos in formation:
    temp_p = players[players['ValueEUR'] < one_player_balance].sort_values(by = f'{pos}Rating', ascending = False).iloc[0]
    while temp_p['ID'] in ids:
        ilc += 1
        temp_p = players[players['ValueEUR'] < one_player_balance].sort_values(by = f'{pos}Rating', ascending = False).iloc[ilc]
    ilc = 0
    ids.append(temp_p['ID'])
    squad.append(temp_p)
    
    #players[players['ValueEUR'] < one_player_balance].sort_values(by = f'{pos}Rating', ascending = False).iloc[0]
```

```python
for id, player in enumerate(squad):
    pos = formation[id]
    print('position:', pos, 'name:', player['Name'], 'overall:', player['Overall'], 'pos_rating:', player[pos+'Rating'], 'age:', player['Age'])
```
